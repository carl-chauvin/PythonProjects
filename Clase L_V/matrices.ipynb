```python
import numpy as np
import matplotlib.pyplot as plt

#Definir la matriz A
A = np.array([[2,1,1],[1,3,2],[1,0,1]])

#Buscar el determinante
detA = np.linalg.det(A)

#Buscar la matriz adjunta
adjA = np.linalg.inv(A).T * detA

#Buscar la matriz transpuesta
transA = A.T 

print('La Matriz cuadrada A es:\n', A)
print('\n El Determinante de la Matriz A es: \n', detA)
print('\n La Matriz Adjunta de la Matriz A es: \n', adjA)
print('\n\n La Matriz Transpuesta de la Matriz A es: \n', transA)
```

```python
#Grafica
plt.imshow(A, cmap='viridis', interpolation='none')
plt.colorbar()
plt.title('Matriz Cuadrada A de 3x3')
plt.show()
```

```python
#Definir la matriz B
B = np.array([[2,-3],[4,6]])

#Buscar el determinante
detB = np.linalg.det(B)

#Buscar la matriz adjunta
adjB = np.linalg.inv(B).T * detB

#Buscar la matriz transpuesta
transB = B.T 

print('La Matriz B es:\n', B)
print('\n El Determinante de la Matriz B es: \n', detB)
print('\n La Matriz Adjunta de la Matriz B es: \n', adjB)
print('\n\n La Matriz Transpuesta de la Matriz B es: \n', transB)
```
